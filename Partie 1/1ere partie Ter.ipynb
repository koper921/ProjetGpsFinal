```python
### retourne classe uniquement de valeur i dans une bd
def df_classF(df,i):
    return  df.loc[df['classe']==i,:]    
```

```python
### fonction principale qui retourne la base de donnée traitée avec les id tous les 100 et toutes les classes
### on précisera que les variables id et y en entrée ne servent à rien dans la fonction

def preparation(df, id, y, length):
    
    rdf = pd.DataFrame() 
    ### c correspond à l'id, on démarre à moins 1 pour le faire en sorte qu'il démarre à 0
    c = -1
    y=[]
    for i in range (1,8):
        
        df_class= df_classF(df ,i)
        ### on split tous les length  et au sein d'une classe de meme valeur
        df_i = np.split(df_class, [k for k in range(length, len (df.loc[df['classe']==i,:])) if k%length==0], axis=0)
        for j in range (0, len(df_i)):
            y.append(i)
            df_i[j]=df_i[j].drop(['classe'], axis=1)
            c=c+1
            df_i[j]['Id']=c
            
                
            rdf=rdf.append(df_i[j])
            
   
    return rdf 
```
